```python
import pandas as pd

df = pd.read_csv("Love Birds.txt", header=None, delimiter="-", names=["Date", "LoveBirds", "Chat1","Chat2"])
df
```

```python
# Extracting the chats from the date column
Chat3 = pd.Series([n if not n[0].isdigit() else None for n in df.Date])

df["Chat"] = Chat3
df.head(20)

nDate = pd.Series([n if n[0].isdigit() else None for n in df.Date])
df["Date"] = nDate
df.head(10)
```

```python
# Cleaning the LoveBirds column
newdf = pd.Series(df["LoveBirds"])

df["Messages"] = newdf.apply(lambda x: str(x).split(":")[-1])
df["LoveBirds"] = df["LoveBirds"].apply(lambda x: str(x).split(":")[0])
df.drop(index = 0, inplace = True)
df.head(20)
df.head()
```

```python
#Stripping the LoveBirds column
df["LoveBirds"] = df.LoveBirds.str.strip()
df.head(20)
```

```python
# Dropping the irrelevant column
df.drop(columns="Chat2", inplace = True)
df.info()
```

```python
df.head()
```

```python
# Merging the chat columns
df["Chat1"].astype("str") + df["Messages"].astype("str")
df["Messages"] =  df["Chat1"].str.cat([df["Messages"], df["Chat"]], sep = "", na_rep="").str.strip("nan")
df.drop(columns = ["Chat1", "Chat"], inplace = True)
df.head()
```

```python
# The dataset will now be saved in a csv file, and be transported to power query for more structuring

df.to_csv("LoveBirds.csv", index=None)
```
